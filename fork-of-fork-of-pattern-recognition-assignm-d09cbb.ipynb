```python
import pandas as pd
import numpy as np
import sklearn 
from sklearn.naive_bayes import GaussianNB 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , accuracy_score , f1_score , recall_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/adult-data/adult.data')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
encoder = OneHotEncoder()
```

```python
x = df.iloc[: , :-1]
y = df[df.columns[-1]]
```

```python
x.head()
```

```python
y.head()
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

```python
y = y.reshape(-1 , 1)
```

```python
x = encoder.fit_transform(x)
y = encoder.fit_transform(y)
```

```python
x = x.toarray()
y = y.toarray()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
model = GaussianNB()
```

```python
y_t = []
for i in range(len(y_train)):
        y_t.append(y_train[i][0])     
```

```python
y_s = []
for i in range(len(y_test)):
    y_s.append(y_test[i][0])
```

```python
y_t = np.array(y_t)
```

```python
print(x_train.shape[0] , y_train.shape[0])
```

```python
print(x_train.shape[0] , y_t.shape[0])
```

```python
model.fit(x_train , y_t)
```

```python
pred = model.predict(x_test)
```

```python
cm = confusion_matrix(y_s , pred)
```

```python
TN = cm[0 , 0]
FP = cm[0 , 1]
FN = cm[1 , 0]
TP = cm[1 , 1]
```

# Accuracy:

```python
accuracy_score(pred , y_s)
```

# Sensitivity:

```python
recall_score(y_s , pred)
```

# Specificity:

```python
print(TN / (TN + FP))
```

# Posterior probability of making over 50k a year:

```python
print(TP /(TP + FP))
```
